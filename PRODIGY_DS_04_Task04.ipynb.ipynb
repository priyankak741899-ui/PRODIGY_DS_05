```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Weather_Condition': ['Clear', 'Rain', 'Fog', 'Clear', 'Rain', 'Cloudy', 'Clear', 'Fog', 'Cloudy', 'Rain'],
    'Road_Condition': ['Dry', 'Wet', 'Wet', 'Dry', 'Wet', 'Dry', 'Dry', 'Wet', 'Dry', 'Wet'],
    'Time_of_Day': ['Morning', 'Evening', 'Night', 'Afternoon', 'Morning', 'Night', 'Evening', 'Night', 'Morning', 'Afternoon'],
    'Latitude': [40.1, 40.2, 40.3, 40.4, 40.5, 40.6, 40.7, 40.8, 40.9, 41.0],
    'Longitude': [-74.1, -74.2, -74.3, -74.4, -74.5, -74.6, -74.7, -74.8, -74.9, -75.0],
    'Contributing_Factor': ['Speeding', 'Rain', 'Poor Visibility', 'Distracted Driving',
                            'Speeding', 'Fatigue', 'Distracted Driving',
                            'Poor Visibility', 'Speeding', 'Rain']
}

df = pd.DataFrame(data)
df.head()
```

```python
sns.countplot(x='Weather_Condition', data=df)
plt.title('Accidents by Weather Condition')
plt.show()
```

```python
sns.countplot(x='Road_Condition', data=df)
plt.title('Accidents by Road Condition')
plt.show()
```

```python
sns.countplot(x='Time_of_Day', data=df)
plt.title('Accidents by Time of Day')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Longitude'], df['Latitude'])
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
sns.countplot(y='Contributing_Factor', data=df)
plt.title('Contributing Factors')
plt.show()
```
